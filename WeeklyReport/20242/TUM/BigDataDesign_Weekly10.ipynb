    - **머신러닝 알고리즘 성능 비교 연구 및 예측:**
        - 주요 연구 내용:
            - 다양한 머신러닝 알고리즘을 선정하여 수집한 농구 데이터를 train set과 test set으로 나누어 예측 성능을 비교·분석함. 각각의 모델은 정확도, 정밀도, F1 점수, RMSE 등으로 평가하였고, 성능 향상을 위해 일부 최적의 파라미터를 재적용하기도 하였음. 특히 수집한 데이터의 특성에 따라 성능이 우수한 알고리즘이 달랐으며, 예를 들어 연속적인 종속 변수(승패 마진)로 설정한 연구에서는 회귀 기반 머신러닝 알고리즘을 사용하여 보다 정밀한 예측을 수행함.
            - 시즌 종료 후 각 팀의 순위를 예측하기 위해, 그 전까지의 데이터를 토대로 각 1~5라운별 로 머신러닝 알고리즘을 적용하여 순위를 예측하고, 정규리그 최종 순위를 예측하였음.
    - **경기 결과 영향 요인 예측 연구:**
        - 주요 연구 내용: 플레이오프 진출 여부 , 경기 결과 등의 예측과 주요한 영향 요인에 대한 분석을 하였음.
        - 영향 요인 분석 방법: 변수 중요도와 SHAP 분석을 통해 주요 변수를 식별하였으며, SHAP은 변수의 중요도와 함께 정(+)·부(-) 관계를 제공하여 변수 중요도의 한계를 보완함.
        - 분석 결과:
            - NBA(플레이오프 진출 여부): SRS(팀의 강도), PA/G(상대팀에 허용한 평균 실점 수), STL(공 스틸 횟수)와 같은 변수가 중요하게 작용함.
            - 여자 농구 월드컵(승패): DREB(상대 슛 실패 시 수비팀의 리바운드 횟수), FG%(필드골 성공률), 턴오버와 같은 변수가 중요하게 작용함.
        - 이러한 연구들은 분석 결과를 통해 긍정적 중요 요인에 대한 능력을 가진 선수 영입 추진, 관련 훈련 강화  등의 전략 수립에 유용하게 활용될 수 있음을 시사함.
    - Insight
        - 기존의 연구를 살펴보며, 머신러닝을 적용한 성능비교 연구는 많이 이루어 지고 있는 것으로 파악함. 최근 들어서, 스포츠의 경기 결과를 예측함과 동시에 결과에 영향을 미치는 주요 요인이 무엇인지 분석하는 연구도 점점 늘어나고 있는 것으로 보임. 따라서, 단순히 머신러닝 분석을 통해 경기결과 예측만 하는 것이 아닌 승패에 어떤 결정적 요인이 영향을 미쳤는가를 설명해줌으로써 훈련전략이나 승리 전략과 같은 제안점을 시사할 수 있다고 생각함.
        - 선행연구들을 토대로 본 프로젝트에서 발전시켜 나갈 수 있는 연구 전략을 몇가지 생각함.
            - 시간에 따른 요인 변화: 경기 초반에는 슛 성공률이 중요할 수 있겠지만, 경기 후반대에는 수비 등이 중요할 수 있지 않을까? 하는 의문점이 들었고, 이는 경기 시간에 따른 특성 중요도 혹은 쿼터 별 맞춤화 전략을 도출하는 방향이 가능성이 있어 보임.
            - 포지션 및 경기 스타일별 분석: 선수 포지션이나 경기 스타일에 따라 중요 요인이 달라질 수 있으므로, 이러한 범주를 나눠 주요 요인을 분석하고, 이를 경기 전략에 적용할 수 있는 방법을 연구를 생각해봄

- **Injury** :
    - 스포츠 분야에서의 injury surveillance 중요성, 부상 관련 데이터 분석이 부상 위험을 줄이기 위한 다양한 조치를 마련하는 데 활용되고 있음을 알게 되었음. 부상 데이터를 활용하여 개발할 수 있는 구체적인 모델에 대해 추가 조사를 진행함.
    - 부상 예측 및 예방을 위해 머신러닝 모델을 구현한 여러 연구를 비교 및 평가하는 연구. 검토된 11개의 연구의 공통점은 모두 classification model에 의존하고 있으며, 주로 부상이 발생했는지 여부와 부상의 종류를 분류하는 데 초점을 맞추고 있음. 언급된 연구 중 가장 성공적으로 적용된 모델을 추가 조사함.
    - 734명의 유소년 축구 선수의 신체 관련 데이터와 XGBoost 알고리즘을 이용해 부상 발생을 예측하고 부상의 유형(overuse or acute)을 분류하는 모델 개발 연구. 예측 정확도는 84%로 상당히 높았으며, 부상 예측에 필요한 중요한 특성을 성공적으로 식별함. 그러나 분석은 각 선수가 경험한 첫 번째 부상에 국한되었으며, 특정 유형 부상에 대한 risk profile의 지속적인 업데이트가 필요하다는 한계점 존재함.
    - Insight:
        - 프로젝트 주제로 생각해볼 수 있는 것:
            - 한 시즌 동안 발생한 여러 부상에 대한 예측 모델 개발.
            - 다양한 연령대, 성별, 스포츠 종목의 부상 데이터를 준비하여 모델의 일반화 향상시키기.
            - 선수들의 부상 이력을 이용한 부상 패턴 분석.
        - 부상 관련 데이터
            - 선수들이 특정 시간에 어떤 부상을 입었는지에 대한 데이터는 수집 가능할 것으로 보임.
            - 그러나 선수들의 신체 측정 데이터, 운동 조정 능력, 체력과 같은 세부적인 건강 정보 수집은 꽤 어려울 것으로 예상됨.
