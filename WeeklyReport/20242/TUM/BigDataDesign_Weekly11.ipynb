- [4] 야구
    - 주제 : 머신러닝 기법을 활용한 야구 경기 승패 예측: 수비와 공격 요소의 상대적 영향력 분석
    → 머신러닝 알고리즘을 활용하여 야구 경기의 승패에 있어 수비 및 공격 데이터의 상대적 중요성을 비교 분석하고자 함. 이를 통해 수비와 공격 요소 중 어느 부분이 경기 결과에 더 큰 영향을 미치는지 밝히고자 함.
    - 데이터 :
        - 수비 관련 지표:
            - Fielding Percentage (F%), Range Factor (RF), Defensive Runs Above, Average (RAA), Errors (E), Assists (Ass), Putouts (PO) 등

        - 공격 관련 지표(타격):
            - AVG(타율), On-base Percentage (OBP), Slugging Percentage (SLG), Home Runs (HR), Runs Batted In (RBI), Walks (BB) 등

        https://statiz.sporki.com/stats/?m=team&m2=batting 
        
    - 모델 : Random Forest와 XGBoost를 활용하여 수비 및 공격 변수를 바탕으로 승패를 예측하고, 변수 중요도를 평가함으로써 경기 결과에 더 중요한 영향을 미치는 요소를 파악하고자 함.
    - 시사점:
        - 수비와 공격 중 어떤 요소가 승패에 더 큰 영향력을 미치는지 상대적인 영향을 알 수 있고, 이를 토대로 팀 전략 수립 및 훈련 방향 설정에 도움 될 것으로 보임.
        - 공격 데이터를 분석할 때, 타격 지표 외에도 주루 및 투구와 관련된 다양한 지표를 변수 중요도나 상관관계 분석을 통해 선정하면, 더 종합적이고 예측력 있는 모델을 구축할 수 있을 것으로 보임.
        - 더 나아가, 위와 같은 연구 내용을 통합하여, 수비, 주루, 타격, 투구 지표를 아우르는 통합적 승패 예측 모델을 연구하는 주제도 가능해 보임.
