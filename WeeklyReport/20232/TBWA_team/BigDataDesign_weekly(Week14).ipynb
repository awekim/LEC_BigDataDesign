    **3. 증감율 계산 예시 식 제공**  
        
    증감율 계산 식은 아래와 같습니다.
    ( ( B - A ) / A ) x 100  

    이때, A는 4월 데이터 값이고 B는 5월 데이터 값입니다. 

    예를 들어, [DA-카카오] 캠페인의 4월 대비 5월 성과를 계산하기 위해서는 먼저 행 레이블에서 같은 카테고리끼리 비교해야 합니다. 4월 데이터에는 행 레이블에 대출 카테고리가 없기 때문에 예금 카테고리를 비교해 보겠습니다.  
    DA-카카오의 4월 CPC는 101이고 5월 CPC는 103이기 때문에, ((103-101)/101)*100은 1.98입니다.
