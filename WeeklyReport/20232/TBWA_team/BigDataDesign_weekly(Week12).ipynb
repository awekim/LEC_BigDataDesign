    4. CPU - Cost per User
    5. 신규 방문 CPU - 신규방문 Cost per User

    **4. 지표에 대한 긍/부정 정리 및 상관관계 정리**  
    
    위 데이터의 변화에서 긍정적인 변화는 다음과 같습니다.  

    긍정
    1. 위의 핵심 10가지 지표는 감소할 수록 긍정 (개선)
    2. 다소 높은 CPC에 비해 CPU 200원대 중반, 심사 CPA 1600원대는 우수한 효율
    3. 심사수 증가는 심사 CPA 개선에 영향
    4. CPC 감소는 CPU 개선에 영향
    5. 예산 높을 수록 긍정
    6. 최대입찰가 감액은 CPU, 신규 방문 CPU 개선에 영향
    7. 심사 CPA 2000원대는 효율이 좋은 편
    8. 유입수, 검색수 증가는 심사 CPA 개선 영향
    
    부정
    1. 3천원대 심사 CPA는 다소 높은 심사 CPA
    2. 광고비 증가는 CPU 증가에 영향 (부정적인 지표), 심사 CPA증가에 영향
    3. 클릭률 감소는 CPU 증가에 영향
    4. 방문자 대비 신규 방문자수 비중 감소는 신규방문 CPU 증가
    5. 유입단가 증가는 전체 CPU 증가에 영향
    6. 노출/ 클릭 감소는 위의 핵심 지표에 부정적인 영향
    7. 클릭률 감소는 CPU, 신규방문 CPU 증가에 영향 (부정적인 지표)
