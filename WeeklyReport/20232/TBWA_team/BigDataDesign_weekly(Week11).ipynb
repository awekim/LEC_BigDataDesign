#### **Multivariate Analysis**
1. 4월 summary_total 상관관계 분석 (heatmap)

    ![4_total_heat](./image/4_total_heat.png)
    
    ![4_total_1](./image/4_total_1.png)
    
    ![4_total_2](./image/4_total_2.png)

    1. CPC (Cost Per Click)

        1. 클릭율, 클릭과 강한 음의 상관관계
        2. 클릭, 클릭율이 CPC 지수와 강한 음의 상관관계를 가진다. CPC는 클릭이 발생할 때만 비용을 지불하는 종량제 광고로, 클릭율이 높을 수록 CPC가 낮아짐을 알 수 있다.  
    <br>

    2. CPS (Cost Per Sales)

        1. 신규방문, 방문자수, 유입수와 강한 음의 상관관계
        2. CPU, 신규방문 CPU와 강한 양의 상관관계
        3. CPS는 매출당 비용을 의미하며, ‘유입수, 방문자수,신규방문,CPU,신규방문 CPU’와 강한 음의 상관관계를 가진다. 유입, 방문, 신규방문,CPU,신규방문 CPU 모두 광고에 대한 매출과 관련이 있는 지수인데, 매출이 높을 수록 (방문이나 유입수가 많을 수록) CPS가 낮아짐을 알 수 있다.  
    <br>

    3. CPU (Cost Per Unique Visitor)

        1. 유입수, 방문자수, 신규방문 강한 음의 상관관계
        2. CPS, 신규방문 CPU 강한 양의 상관관계
        3. CPU는 신규 방문수에 대한 비용을 의미하며,  ‘유입수, 방문자수, 신규방문, CPS, 신규방문 CPU’와 강한 음의 상관관계를 가진다. 해당 지표 모두 신규 방문자와 관련이 있는 지표로, 해당 지표들이 높을 수록 CPU가 낮아짐을 알 수 있다.  
    <br>

    4. 신규방문CPU

        1. 유입수, 방문자수, 신규방문 강한 음의 상관관계
        2. CPS, CPU 강한 양의 상관관계
        3. 신규방문 CPU는 신규로 방문한 Unique visitor에 대한 비용으로, 유입수, 방문자수, 신규방문 수가 높을 수록 신규방문 CPU는 낮아진다. 반면에, CPS와 CPU가 높을 수록 신규방문 CPU 또한 높아진다.  
    <br>

    5. 접수 CPA (Cost Per Action)

        1. 심사CPA와 강한 양의 상관관계
        2. 접수 CPA는 접수 행동에 대한 비용을 의미하며, 심사 CPA는 심사 행동에 대한 비용을 의미한다. 접수  CPA가 높을 수록 심사 CPA 또한 높아짐을 알 수 있다.  
    <br>

    6. 심사 CPA (Cost Per Action)

        1. 접수 CPA 와 강한 양의 상관관계
        2. 접수 CPA는 접수 행동에 대한 비용을 의미하며, 심사 CPA는 심사 행동에 대한 비용을 의미한다. 접수  CPA가 높을 수록 심사 CPA 또한 높아짐을 알 수 있다.  
    <br>

    7. 승인 CPA

        1. 승인수와 강한 음의 상관관계
        2. 승인 CPA는 승인 행동에 대한 비용을 의미하며, 승인수와 강한 음의 상관관계를 가진다. 승인수가 높을 수록 승인 CPA가 낮아짐을 알 수 있다.  
    <br>

    8. CPA (Cost Per Action)

        1. 예금+대출율, 예금+ 대출, 대출과 강한 음의 상관관계
        2. 대출 CPA와 강한 양의 상관관계
        3. CPA는 행동(액션)당 비용을 의미하며, ‘예금+대출율, 예금+대출, 대출, 대출 CPA’와 강한 상관관계를 가지는 걸 알 수 있는데, 예금+대출율과 대출수가 높을 수록 CPA는 낮아지며, 대출 CPA가 높을 수록 CPA가 높아짐을 알 수 있다.  
    <br>

    9. 예금 CPA (Cost Per Action)

        1. 예금과 강한 음의 상관관계
        2. 예금 CPA는 예금 행동당 비용을 의미하며, 예금이 높을 수록 예금 CPA가 낮아짐을 알 수 있다.  
    <br>

    10. 대출 CPA (Cost Per Action)
    
        1. 예금+대출 , 대출과 강한 음의 상관관계
        2. CPA 와 강한 양의 상관관계
        3. 대출 CPA는 대출 행위 당 비용을 의미하며, ‘예금+대출, 대출, CPA’와 강한 상관관계를 가지는데, 예금+대출, 대출의 수가 높을 수록 대출 CPA가 낮아지며, CPA가 높을 수록 대출 CPA가 높아짐을 알 수 있다.
