### Part 1. Brief background of Random Forest

#### **The situation before this methodology was introduced:**
  
Before the introduction of the Random Forest model, Decision Trees were predominantly used. However, Decision Trees had several issues. 

First, there was a problem of overfitting. Decision Trees tend to classify the training data perfectly, which can lead to excessively deep and complex trees that learn even the noise in the training data, resulting in complex rules. While this leads to high performance on training data, it causes a decline in performance on test data due to overfitting. 

Second, there was instability, where even small changes in the data could significantly alter the results. Changes in the split criterion at the upper node can modify the entire structure below, leading to decreased stability. This indicates that the model is highly sensitive to the data and can produce inconsistent and large variations in prediction results. 

Third, although bagging was used, there were limitations in performance improvement due to high correlation between individual trees. Bagging is a method that trains several trees on different data samples and averages the results to improve performance. The goal is to reduce overfitting and enhance prediction stability by combining the predictions of individual trees. However, if the individual trees are highly correlated (meaning the trees are not independent from each other), the benefit of aggregating their predictions is limited.

---

#### **Why this methodology is introduced:**

To address the problems associated with Decision Trees, Random Forest was introduced. Random Forest is an ensemble method that improves predictive performance by combining multiple decision trees. Each tree is trained independently, and the results are aggregated to make a final prediction, thereby compensating for the weaknesses of decision trees. 

First, to solve the overfitting problem, Random Forest employs ensemble learning. By training multiple trees and aggregating their predictions using averaging or majority voting, it prevents individual trees from fitting excessively to the training data. By combining several trees, more generalized results can be obtained, thus mitigating overfitting issues. 

Second, the problem of large variations in results due to small changes in data is addressed by using different data samples to train each tree. Since each tree is trained on different data, the instability of individual trees has a reduced impact on the overall model, resulting in more consistent prediction outcomes. 

Third, the correlation problem is resolved. To overcome the limitations of bagging, Random Forest adds feature randomness during the training of trees. This involves randomly selecting variables to use during the training of each tree, helping to lower the correlation between trees. As a result, the trees can make independent predictions, leading to greater performance improvement.

### Part 2. Key concept of methodology
#### **The key concept of the methodology with mathematical equations:**

1. **Ensemble Prediction**

    Random Forest is an ensemble model composed of multiple decision trees, combining the predictions from each tree to create a final prediction.

    - Ensemble Prediction for Regression Problems: The final prediction is generated by averaging the predictions from each tree.
    - 
    $$\hat{y} = \frac{1}{T} \sum_{t=1}^{T} h_t(x)$$
    

    - ùëá = Number of trees
    
    - h_t(x) Prediction from tree ùë°


    This formula calculates the final predicted value ùë¶^ by averaging the predictions from all trees.
    
    
    - Ensemble Prediction for Classification Problems: The final prediction is determined by the most frequently occurring value (majority vote) from the predictions of all trees.
    - 
    $$\hat{y} = \text{mode} \{ h_1(x), h_2(x), ..., h_T(x) \}$$

    - The final classification is decided by the majority vote among the predictions h_t(x) from each tree.

2. **Generalization Error**

    The generalization error of Random Forest depends on the error rates of the individual trees, which are reduced by the ensemble of models with low correlation. The generalization error of Random Forest can be expressed in terms of correlation ùúå and the accuracy ùë† of each tree.

    $$GE \approx \rho (1 - s^2)$$

   - œÅ = Correlation between trees

   - s = Accuracy of each tree
  
   - The generalization error decreases as the correlation between trees is low and the accuracy of each tree is high.

3. **Variable Importance**

    In Random Forest, the importance of each feature is calculated based on how often the feature is used to split nodes in the model and the reduction in impurity achieved through those splits. The greater the impurity reduction, the higher the importance of the feature.

    Variable Importance for a Specific Tree:
    $$f_i = \sum_{j \in \text{nodes}} \frac{w_j \Delta C_j}{\text{Total Weight}}$$

    - ùëì_ùëñ = Importance of feature 
    - ùë§_ùëó = Number of samples reaching node 
    - Œîùê∂_ùëó = = Reduction in impurity at node 
    - This formula calculates the importance of each feature by weighting the impurity reduction at each node.



    Variable Importance for the Entire Random Forest:
    $$RFf_i = \frac{1}{T} \sum_{t=1}^{T} f_i^{(t)}$$

    - RFf_ùëñ = Importance of feature ùëñ across the entire Random Forest
    - ùëá = Number of trees
    - ùëì_ùëñ^(t) = Importance of feature ùëñ in tree ùë°
    - This formula averages the feature importances computed in each tree to obtain the final feature importance.

#### **The main strength of the Random Forest:**


Random Forest uses an ensemble method composed of multiple decision trees, significantly enhancing performance. Here are the key strengths of this methodology:

1. Prevention of Overfitting
    Random Forest effectively reduces the overfitting problem because each tree is trained independently, and the final prediction is made by combining the results from multiple trees. While a single decision tree tends to overfit the training data, Random Forest prevents overfitting by averaging or voting on the predictions from several trees, providing a more generalized performance.

2. Stability and Consistency
    Random Forest maintains stable performance even with small changes in the training data. Decision trees can be sensitive to training data, resulting in significant changes in the tree structure and prediction results with slight modifications. However, Random Forest mitigates this issue by combining predictions from multiple trees, making it less sensitive to data changes and ensuring stability and consistency.

3. Capability to Handle Diverse Data
    Random Forest excels at processing high-dimensional data, multi-variable data, and complex nonlinear relationships. Since different trees learn using various data samples and features, they effectively capture diverse patterns in the data. It is particularly effective in scenarios with high feature interactions and noisy data.

4. Evaluation of Feature Importance
    Random Forest can evaluate the importance of each feature, helping to understand which features the model relies on most. This plays a crucial role in tasks such as feature selection. Random Forest provides automatic importance scores for features, enhancing interpretability.

5. Suitability for Parallel Processing
    Random Forest is suitable for parallel processing because individual trees are trained independently. There are no interactions between trees, allowing for simultaneous training of multiple trees, significantly improving training speed. It can efficiently handle large datasets.

6. Applicability to a Variety of Problems
    Random Forest is effective for both classification and regression problems. It achieves high performance for both types through majority voting and averaging methods, making it widely used across various fields. For instance, it is employed in finance, healthcare, marketing, and manufacturing for predictive tasks.

### Part 3. Example

#### **Using simple toy data**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
cancer = load_breast_cancer()
print(cancer.keys())
```

```python
print(cancer.target_names)
print(cancer.feature_names)
```

```python
# Split the training and testing data in an 8:2 ratio for learning and evaluation
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=42 )
```

```python
# Use entropy for splitting the trees
# bootstrap=True: Activate bagging to allow duplication when sampling data
# max_depth=5: Limit the depth of each tree to prevent overfitting
rand_clf = RandomForestClassifier(criterion='entropy', bootstrap=True, random_state=42, max_depth=5)


# Model training and evaluation
rand_clf.fit(X_train, y_train)
y_pred = rand_clf.predict(X_test)

print('Training set accuracy: {:.3f}' .format(rand_clf.score(X_train, y_train)))
print('Testing set accuracy: {:.3f}' .format(rand_clf.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# confusion matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Extract feature importance

for name, value in zip(cancer.feature_names , rand_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))

# Visualization settings
plt.figure(figsize=(10, 8)) 
sns.set(style="whitegrid")  

# Set color palette 
palette = sns.color_palette("coolwarm", len(cancer.feature_names))

# Visualize feature importance by column
ax = sns.barplot(x=rand_clf.feature_importances_, y=cancer.feature_names, palette=palette)

plt.title("Feature Importance", fontsize=16, fontweight='bold')

plt.xlabel("Importance", fontsize=14)
plt.ylabel("Features", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

sns.despine(left=True, bottom=True)

for i, v in enumerate(rand_clf.feature_importances_):
    ax.text(v + 0.01, i, f'{v:.3f}', color='black', va='center', fontsize=12)

plt.tight_layout()
plt.show()
```
